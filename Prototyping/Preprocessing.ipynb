```python
from google.colab import drive
import pandas as pd
import spacy
import regex as re
import nltk
from nltk.corpus import stopwords
from tqdm.notebook import tqdm
nltk.download('stopwords')
```

# Preprocessing of email text

```python
drive.mount("/content/drive")
```

```python
file = open('/content/drive/MyDrive/Mails.csv',mode='r', encoding='utf-8')
lines = file.read().split('!#ENDOFLINE#!')
file.close()

# Split the data
for i in range(len(lines)):
    lines[i] = lines[i].split('!#!')
    for j in range(len(lines[i])):
        lines[i][j] = lines[i][j].strip()
        
# Load it to a data frame
df = pd.DataFrame(lines)
df.columns = df.iloc[0]
df = df[1:]

# Small bugfix (work around)
df = df.rename(columns={"ï»¿Subject" : "Subject"})
```

```python
nlp = spacy.load('en_core_web_sm')
COLUMN = ['Body']
#symbols = re.compile('[/(){}\[\]\|@,;:]')
#symbols = ['<', '>', '|', ':', '/', '(', ')', '[', ']', ',', '.', '\\', ' ', '_', '-', '\'', '\"', '!', '?', ';', '&']
STOPWORDS = set(stopwords.words('english'))

df["Body_processed"] = ""


def prepare(data_frame):
    for index, row in tqdm(data_frame.iterrows(), total=data_frame.shape[0]):
        for x in COLUMN:
            if not pd.isnull(data_frame[x][index]):
                # Add title to body
                text = row['Subject'] + " " + row["SenderName"] + " " + row["CCAddresses"] + " " + row['Body']
                # Lemmatize
                doc = nlp(text)
                result = []
                for token in doc:
                    str_token = str(token)
                    if not (str_token.startswith("http://") or str_token.startswith("https://") or len(str_token.strip()) <= 1 or '\\n' in str_token or '\n' in str_token):
                        lemma = token.lemma_.lower()
                        if not lemma in STOPWORDS:
                            result.append(lemma)
                
                result.append(row["SenderAddress"])
                data_frame['Body_processed'][index] = result
    return data_frame
   
df = prepare(df)
df.to_csv('/content/drive/MyDrive/Mails_prepared.csv')
```
