## LOOP for every cluster

```python
from gensim import corpora
from gensim import corpora
import gensim
from pprint import pprint

# get all columns 
print({*df})
```

```python
cluster = 'agglo'

clusters = df[cluster].unique()
df['topic'] = 'none'
print(clusters)

for x in clusters:
  df_filtered = df
  df_filtered = df_filtered.loc[df[cluster] == x]  
  texts = df_filtered['Body_processed'].to_numpy()

  # create dictionary 
  dictionary = corpora.Dictionary(texts)
  
  # create BOW
  corpus = [dictionary.doc2bow(text) for text in texts]
  
  # number of topics
  num_topics = 3
  
  # Build LDA model
  lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                    id2word=dictionary,
                                    num_topics=num_topics)

  # Print the Keyword
  print(dictionary[(lda_model.get_topic_terms(0,1)[0][0])])
  df.loc[df.cluster == x, 'topic'] = dictionary[(lda_model.get_topic_terms(0,1)[0][0])]

df.to_csv('/content/drive/MyDrive/Mails_topics.csv')
```
